```python
def plot_multigraph(cpu_data,gpu_t4_data, gpu_v100_data, gpu_a100_data, title="undefined", xlabel = "xlabel", ylabel= "ylabel"):
  plt.plot(x,cpu_data,'bo--', label= "CPU")
  plt.plot(x,gpu_t4_data,'ro--', label = "GPU T4")
  plt.plot(x,gpu_v100_data,'go--', label = "GPU V100")
  plt.plot(x,gpu_a100_data,'yo--', label = "GPU A100")

  plt.legend()
  plt.title(label = "RockPool:{}".format(title))
  plt.xticks(ticks=x)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.show()
```

```python
plot_multigraph(cpu_acc, gpu_t4_acc, gpu_v100_acc, gpu_a100_acc, title = "accuracy per training epoch", xlabel = "epoch", ylabel="accuracy %")
```
